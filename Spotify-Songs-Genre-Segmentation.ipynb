```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
scaler=StandardScaler()
num_cols = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms']
df[num_cols]=scaler.fit_transform(df[num_cols])
while True:
    print("options")
    print("1. show Correlation Matrix")
    print("2. Cluster by Playlist Genres")
    print("3. Bar Chart for Top Listened Artists")
    print("4. Bar Chart for Top Listened Playlists")
    print("5. Exit")
    choice = input("Enter your choice (1/2/3/4/5): ")
    
    if choice == '1':
        correlation_matrix = df[num_cols].corr()
        plt.figure(figsize=(12, 8))
        sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',fmt=".2f")
        plt.title('Correlation Matrix')
        plt.show()
    
    elif choice == '2':
        genres = df['playlist_genre'].unique()
        for genre in genres:
            genre_data = df[df['playlist_genre'] == genre]
            kmeans = KMeans(n_clusters=3, random_state=42)
            genre_data['cluster'] =kmeans.fit_predict(genre_data[num_cols])
            plt.figure(figsize=(8, 6))
            sns.scatterplot(x='danceability', y='energy', hue='cluster',data=genre_data)
            plt.title(f'Clustering for {genre}')
            plt.show()
    elif choice == '3':
        top_artists = df['track_album_name'].value_counts().head(10)
        plt.figure(figsize=(10, 6))
        top_artists.plot(kind='barh')
        plt.xlabel('Number of Listens')
        plt.ylabel('Artist')
        plt.title('Top Listened Artists')
        plt.gca().invert_yaxis()
        plt.show()
    elif choice == '4':
        top_playlists = df['playlist_name'].value_counts().head(10)
        plt.figure(figsize=(10, 6))
        top_playlists.plot(kind='barh')
        plt.xlabel('Number of Listens')
        plt.ylabel('Playlist Name')
        plt.title('Top Listened Playlists')
        plt.gca().invert_yaxis()
        plt.show()
    elif choice == '5':
        break
    else:
        print("Invalid choice. Please enter a valid option (1/2/3/4/5).")
```
